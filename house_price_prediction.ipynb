```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/housing.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
X= df.drop(['median_house_value'] ,axis=1)
y= df['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

# Data visualization and exploration

```python
train_data.hist(figsize=(10,8))
```

```python
train_data.head()
```

```python
new_df = train_data.drop(['ocean_proximity'], axis =1)
new_df.head()
```

```python

plt.figure(figsize=(12,8))
sns.heatmap(new_df.corr(), annot=True ,cmap='YlGnBu')
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(10,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
td1=train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
td1
```

```python

plt.figure(figsize=(12,8))
sns.heatmap(td1.corr(), annot=True ,cmap='YlGnBu',cbar=False)
```
